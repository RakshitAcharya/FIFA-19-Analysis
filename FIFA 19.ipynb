```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.drop(["ID","Special"],axis = 1,inplace = True)
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0','Crossing', 'Finishing', 'HeadingAccuracy', 'Volleys','Real Face','Dribbling', 'Curve', 'FKAccuracy', 'BallControl','ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving',
       'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes'],axis =1 , inplace  = True)
```

```python
df.head()
```

```python
df["SPEED"] = (df["Acceleration"]+df["SprintSpeed"]+0.75*df["Agility"]) /3
df.drop(["Acceleration","SprintSpeed","Agility"],axis =1 , inplace = True)
```

```python
df["PASSING"] = (df["ShortPassing"]+df["LongPassing"]+df["Reactions"]+df["Balance"]) /4
df.drop(["ShortPassing","LongPassing","Reactions","International Reputation","Jersey Number","Balance","Contract Valid Until"],axis =1 , inplace = True)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df2 = df.dropna(subset = ['Club',"Preferred Foot"])
df2.isna().sum()
```

```python
df2['Release Clause'] = df2['Release Clause'].fillna(0)
```

## EDA


### 1. PREFFERED FOOT

```python
ax = sns.countplot(x = "Preferred Foot", data = df2);
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), ha='center', va='top', color='white', size=18)
plt.title("NUMBER OF LEFT/RIGHT FOOT PLAYERS")
plt.show()
```

### 2. DISTRIBUTION OF PLAYERS NATIONALITIES


```python
df3 = pd.DataFrame(df2.groupby("Nationality")["Name"].count())
df3 = df3[df3["Name"]>300]
df3.rename(columns = {"Name":"Nationality_count"},inplace = True)
df3.reset_index(inplace = True)
```

```python
plt.figure(figsize =(10,6))
sns.barplot(x= "Nationality_count", 
            y= "Nationality",
            data = df3,
            order=df3.sort_values('Nationality_count',ascending = False).Nationality
           );
plt.title("Distribution of Nationality")
plt.yticks(rotation = 25)
plt.show()
```

### 3. AVERAGE RATING OF PLAYERS ( per position ) 


```python
df2["Position"].unique()
```

```python
df3 = pd.DataFrame(df2.groupby("Position")["Overall"].mean())
df3.reset_index(inplace =True)

plt.figure(figsize =(12,6))
sns.barplot(x ="Overall",
            y= "Position",
            data = df3,
            order = df3.sort_values('Overall', ascending = False).Position
           );
plt.title("Average Ratings For Positions")
plt.yticks(rotation = 25)
plt.show()
```

### 4. AVERAGE RATING OF PLAYERS IN BIG TEAMS

```python
clubs = ['FC Barcelona', 'Juventus', 'Paris Saint-Germain',
       'Manchester United', 'Manchester City', 'Chelsea', 'Real Madrid',
       'Atl�tico Madrid', 'FC Bayern M�nchen', 'Tottenham Hotspur',
       'Liverpool', 'Arsenal', 'Milan', 'Inter',
       'Borussia Dortmund']
df3 = df2.loc[df2['Club'].isin(clubs)]
df3 = df3[df3["Overall"]>75]

df3 = pd.DataFrame(df3.groupby("Club")["Overall"].mean())
df3.reset_index(inplace =True)

ax =sns.barplot(x="Club",
           y="Overall",
           data = df3,
           order = df3.sort_values("Overall", ascending =False).Club);
for p in ax.patches:
    x = p.get_height()
    ax.annotate(f'\n{round(x,1)}', (p.get_x()+0.35, p.get_height()), ha='center', va='top', color='black', size=8)

plt.title("Average Rating of Big Teams")
plt.xticks(rotation = 90 )
plt.show()
```

### 9. TOP PLAYERS FOR POSITIONS

```python
GK_position = df2[df2["Position"]=="GK"].sort_values("Overall",ascending=False)[["Name","Overall"]]
GK_position = GK_position.iloc[:10,:]

ST_position = df2[df2["Position"]=="ST"].sort_values("Overall",ascending=False)[["Name","Overall"]]
ST_position = ST_position.iloc[:10,:]

LW_position = df2[df2["Position"]=="LW"].sort_values("Overall",ascending=False)[["Name","Overall"]]
LW_position = LW_position.iloc[:10,:]

CM_position = df2[df2["Position"]=="RCM"].sort_values("Overall",ascending=False)[["Name","Overall"]]
CM_position = CM_position.iloc[:10,:]
```

```python
arr = [GK_position,ST_position,LW_position, CM_position]
arri =  ["GK","ST","LW","CM"]
for i in range(len(arr)):
    ax = sns.barplot(x="Name",
               y = "Overall",
               data = arr[i] )
    for p in ax.patches:
        x = p.get_height()
        ax.annotate(f'\n{x}', (p.get_x()+0.35, p.get_height()), ha='center', va='top', color='black', size=8)

    plt.title("Top players in the position :{}".format(arri[i]))
    plt.xticks(rotation = 90)
    plt.show()
```
